```python
def read_df(path: str, sep: str = None, encoding: str = None) -> pd.DataFrame:
    """
    Читает файл CSV и возвращает его содержимое в виде датафрейма.
    :param path: путь к папке, содержащей файлы CSV
    :param sep: опциональный разделитель столбцов
    :param encoding: опциональная кодировка файла
    :return: датафрейм
    """
    df = pd.read_csv(path,
                     sep=sep,
                     encoding=encoding)
    return df
```

```python
def create_histogram(df: pd.DataFrame,
                     x: str,
                     title=None,
                     xlabel: str = None,
                     ylabel: str = None,
                     figsize: tuple = (9, 7),
                     kde: bool = False) -> None:
    """
    Создает гистограмму количества квартир в зависимости от этажности дома.
    :param df: исходный датафрейм
    :param x: имя столбца по x
    :param title: заголовок гистограммы
    :param xlabel: название оси x
    :param ylabel: название оси y
    :param figsize: размеры фигуры
    :param kde: флаг указывающий на примененние параметра kde
    :return: None
    """

    plt.figure(figsize=figsize)

    sns.histplot(data=df, x=x, color='purple', kde=kde)
    plt.title(title, fontsize=15, pad=10)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)

    plt.show()
```

```python
def remove_outliers(df: pd.DataFrame, column: str, threshold: float = 1.5) -> pd.DataFrame:
    """
    Удаляет выбросы с правой стороны
    :param df: датафрейм
    :param column: колонка, по которой будем удалять выбросы
    :param threshold: коэфициент, показывающий сколько будем удалять
    :return: датафрейм
    """
        
    data = df[column].values
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    upper_bound = q3 + threshold * iqr

    filtered_df = df[df[column] <= upper_bound]
    return filtered_df
```

```python
def check_normality(data: pd.DataFrame, alpha: float = 0.05) -> None:
    """
    Проверяет нормальность распределения данных с помощью теста Шапиро-Уилка.
    :param data: данные для проведения теста на нормальность
    :param alpha: уровень значимости
    :return: None
    """

    test_shapiro = stats.shapiro(data)

    if test_shapiro.pvalue < alpha:
        print('Распределение не нормальное')
    else:
        print('Распределение нормальное')
```

```python
def split_data(data: pd.DataFrame, target: str, log: bool = False):
    """
    Разделяет данные на матрицу признаков и векторы целевых переменных.
    :param data: pd.DataFrame, исходные данные
    :param target_columns: list, список имен столбцов с целевыми переменными
    :param log: флаг показывающий нужно ли логарифмировать целевую переменую
    :return: матрица признаков X
    :return: целевая переменая
    """

    X = data.drop(columns=target, axis=1)
    y = data[target].values
    if log == True:
        y_log = np.log(data[target]).values
        return X, y, y_log
    else:
        return X, y
```

```python
def prepare_data(X: pd.DataFrame,
                 y: pd.DataFrame,
                 y_log: pd.Series = None,
                 test_size: float =0.2,
                 random_state: int=None,
                 log: bool = False):
    """
    Подготавливает данные для обучения модели.
    :param X: матрица признаков
    :param y: вектор целевой переменной (обычная шкала)
    :param y_log:  вектор целевой переменной (логарифмическая шкала)
    :param test_size: размер тестового набора
    :param random_state: значение для генерации случайной последовательности
    :return:X_train_bin: матрица признаков для обучения с бинаризацией категориальных признаков
    :return:X_test_bin: матрица признаков для тестирования с бинаризацией категориальных признаков
    :return:X_train_log: матрица признаков для обучения с масштабированием логарифмических признаков
    :return:X_test_log: матрица признаков для тестирования с масштабированием логарифмических признаков
    :return:y_train: вектор обучающей выборки целевой переменной (обычная шкала)
    :return:y_test: вектор тестовой выборки целевой переменной (обычная шкала)
    :return:y_train_log: вектор обучающей выборки целевой переменной (логарифмическая шкала)
    :return:y_test_log: вектор тестовой выборки целевой переменной (логарифмическая шкала)
    :return:y_test_exp: вектор экспоненциальных значений тестовой выборки (восстановление из логарифмической шкалы)
    :return:y_train_exp: вектор экспоненциальных значений обучающей выборки (восстановление из логарифмической шкалы)
    :return:X_train_std: стандартизированная матрица признаков для обучения
    :return:X_test_std: стандартизированная матрица признаков для тестирования
    :return: X_train: матрица признаков для обучения
    :return: X_test: матрица признаков для тестирования
    :return: cat_features: список с категориальнными признаами
    """
    # с бинаризацией
    X_train_bin, X_test_bin, y_train, y_test = train_test_split(
        pd.get_dummies(X), y, test_size=test_size, random_state=random_state)

    # без бинаризации
    X[X.select_dtypes('object').columns] = X[X.select_dtypes(
        'object').columns].astype('category')

    X_train = X.loc[X_train_bin.index]
    X_test = X.loc[X_test_bin.index]

    # список с категориальнными признаами
    cat_features = X.select_dtypes('category').columns.tolist()

    if log == True:
        X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
            pd.get_dummies(X),
            y_log,
            test_size=test_size,
            random_state=random_state)

        y_test_exp = np.exp(y_test_log)
        y_train_exp = np.exp(y_train_log)

        mms = MinMaxScaler()
        X_train_std = mms.fit_transform(X_train_log)
        X_test_std = mms.transform(X_test_log)

        return X_train_bin, X_test_bin, y_train, y_test, X_train, X_test, cat_features, \
        X_train_log, X_test_log, y_train_log, y_test_log, y_test_exp, y_train_exp, \
        X_train_std, X_test_std
    return X_train_bin, X_test_bin, y_train, y_test, X_train, X_test, cat_features        
```

```python
def check_overfitting_df(model, X_train: pd.DataFrame, y_train: pd.Series,
                         X_test: pd.DataFrame, y_test: pd.Series, metric_fun,
                         name: str) -> pd.DataFrame:
    """
    Проверяет наличие overfitting модели и добавляет результаты в датафрейм.

    Аргументы:
    :param model: обученная модель
    :param X_train: матрица признаков для обучения
    :param y_train: вектор целевой переменной для обучения
    :param X_test: матрица признаков для тестирования
    :param y_test: вектор целевой переменной для тестирования
    :param metric_fun: метрика для оценки качества модели
    :param name: имя модели
    :return: датаврейм
    """
    df_check_overfitting = pd.DataFrame()

    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    value_train = metric_fun(y_train, y_pred_train)
    value_test = metric_fun(y_test, y_pred_test)

    df_check_overfitting['model'] = [name]

    df_check_overfitting['metric'] = [metric_fun.__name__]
    df_check_overfitting['value_train'] = [value_train]
    df_check_overfitting['value_test'] = [value_test]
    df_check_overfitting['overfitting'] = [
        abs(value_train - value_test) / value_test * 100
    ]

    return df_check_overfitting
```

```python
def ridge_model(X_train: np.ndarray,
                y_train: np.ndarray,
                X_test: np.ndarray,
                log: bool = False,
                random_state: int = 10) -> np.array:
    """
    Предсказывает значения с помощью модели Ridge
    :param X_train: матрица признаков для обучения
    :param y_train: вектор целевой переменной для обучения
    :param X_test: матрица признаков для тестирования
    :param log: флаг, указывающий на то были ли логарифмированы данные
    :param random_state: значение для генерации случайной последовательности
    :return: массив с предсказанными значениями
    """
    ridge = Ridge(random_state=random_state)
    ridge.fit(X_train, y_train)
    r_pred = ridge.predict(X_test)
    if log == True:
        r_pred = np.exp(r_pred)
        max_value = np.max(r_pred[np.isfinite(r_pred)])
        r_pred[~np.isfinite(r_pred)] = max_value

    return r_pred, ridge
```

## XGBoost

```python
def split_train_val(X: pd.DataFrame,
                    y: pd.Series,
                    X_bin: pd.DataFrame,
                    test_size: float = 0.16,
                    shuffle: bool = True,
                    random_state: int = None,
                    bins: bool = False):
    """
    Разделяет данные на обучающий и валидационный наборы.

    Аргументы:
    :param X: матрица признаков
    :param y: вектор целевой переменной
    :param X_bin: матрица признаков с бинаризацией
    :param test_size: размер валидационного набора
    :param shuffle: флаг перемешивания данных перед разделением
    :param random_state: значение для генерации случайной последовательности
    :param bin: флаг бинаризации
    :return: X_train_: матрица признаков для обучения
    :return: X_val: матрица признаков для валидации
    :return: y_train_: вектор обучающей выборки целевой переменной
    :return: y_val: вектор валидационной выборки целевой переменной
    :return: X_train_bin: матрица признаков для обучения биноризованая
    :return: X_val_bin: матрица признаков для валидации биноризованая
    """

    X_train_, X_val, y_train_, y_val = train_test_split(
        X, y, test_size=test_size, shuffle=shuffle, random_state=random_state)
    eval_set = [(X_val, y_val)]
    if bins == True:
        X_train_bin_, X_val_bin, y_train_, y_val = train_test_split(
            X_bin,
            y,
            test_size=test_size,
            shuffle=shuffle,
            random_state=random_state)
        eval_set_bin = [(X_val_bin, y_val)]
        return X_train_, X_train_bin_, X_val, X_val_bin, y_train_, y_val, eval_set, eval_set_bin

    return X_train_, X_val, y_train_, y_val, eval_set
```

```python
def XGBoost_model(X_train: pd.DataFrame,
                  y_train: pd.Series,
                  X_test: pd.DataFrame,
                  eval_set,
                  random_state: int = 10,
                  eval_metric: str = 'mae',
                  early_stopping_rounds: int = 100):
    """
    Обучает модель XGBoost на обучающих данных и делает предсказание на тестовых данных.
    :param X_train: матрица признаков для обучения
    :param y_train: вектор целевой переменной для обучения
    :param X_test: матрица признаков для тестирования
    :param random_state: значение для генерации случайной последовательности
    :param eval_metric: метрика
    :param early_stopping_rounds: ранняя остановка
    :return: массив с предсказанными значениями и обученую модель
    """

    clf = XGBRegressor(random_state=random_state)

    clf.fit(X_train,
            y_train,
            eval_metric=eval_metric,
            eval_set=eval_set,
            early_stopping_rounds=early_stopping_rounds,
            verbose=0)

    y_pred = clf.predict(X_test)

    return y_pred, clf
```

## LGBM

```python
def lgbm_model(X_train: pd.DataFrame,
               y_train: pd.Series,
               X_test: pd.DataFrame,
               eval_set: list,
               random_state: int = 10,
               objective: str = "mae",
               eval_metric: str = "mae",
               early_stopping_rounds: int = 100):
    """
    Обучает модель LGBMRegressor на обучающих данных, делает предсказание на тестовых данных
    и возвращает предсказанные значения и обученную модель.
    :param X_train: матрица признаков для обучения
    :param y_train: вектор целевой переменной для обучения
    :param X_val: матрица признаков для валидации
    :param y_val: вектор целевой переменной для валидации
    :param X_test: матрица признаков для тестирования
    :param eval_set: список с парами (X_val, y_val) для оценки качества модели на валидации
    :param objective: функция потерь для оптимизации 
    :param random_state: значение для генерации случайной последовательности 
    :param eval_metric: метрика для оценки качества модели 
    :param early_stopping_rounds: число итераций для ранней остановки 
    :return: массив с предсказанными значениями и обученую модель
    """

    lgbm = LGBMRegressor(random_state=random_state, objective=objective)

    lgbm.fit(X_train,
             y_train,
             eval_metric=eval_metric,
             eval_set=eval_set,
             verbose=False,
             early_stopping_rounds=early_stopping_rounds)

    y_pred_lgbm = lgbm.predict(X_test)

    return y_pred_lgbm, lgbm
```

## CatBoost

```python
def catboost_model(X_train: pd.DataFrame,
                   y_train: pd.Series,
                   X_test: pd.DataFrame,
                   eval_set: list,
                   cat_features: list,
                   random_state: int = 10,
                   eval_metric: str = "MAE"):
    """
    Обучает модель CatBoostRegressor на обучающих данных и делает предсказание на тестовых данных.
    :param X_train: pd.DataFrame, матрица признаков для обучения
    :param y_train: pd.Series or array-like, вектор целевой переменной для обучения
    :param X_test: pd.DataFrame, матрица признаков для тестирования
    :param eval_set: список с парами (X_val, y_val) для оценки качества модели на валидации
    :param cat_features: list, список категориальных признаков
    :param random_state: значение для генерации случайной последовательности 
    :param eval_metric: метрика для оценки качества модели 
    :return: массив с предсказанными значениями и обученую модель
    """

    ctb = CatBoostRegressor(allow_writing_files=False,
                            random_state=random_state,
                            eval_metric=eval_metric,
                            cat_features=cat_features)

    ctb.fit(X_train, y_train, eval_set=eval_set, verbose=False)

    y_pred = ctb.predict(X_test)

    return y_pred, ctb
```

# Tuning

```python
def objective_ctb(trial, X, y, N_FOLDS, random_state, cat_feat):
    params = {
        "n_estimators": trial.suggest_categorical("n_estimators", [300]),
        "learning_rate": trial.suggest_categorical("learning_rate", [0.2945510204081633]),
        "max_depth": trial.suggest_int("max_depth", 3, 12),
        "l2_leaf_reg": trial.suggest_uniform("l2_leaf_reg", 1e-5, 1e2),
        'border_count': trial.suggest_categorical('border_count', [128, 254]),
        "random_state": random_state
    }

    #if params["bootstrap_type"] == "Bayesian":
    #    params["bagging_temperature"] = trial.suggest_float(
   #         "bagging_temperature", 0, 100)
   # elif params["bootstrap_type"] == "Bernoulli":
    #    params["subsample"] = trial.suggest_float(
    #        "subsample", 0.1, 1, log=True)

    cv = KFold(n_splits=N_FOLDS, shuffle=True, random_state=random_state)

    cv_predicts = np.empty(N_FOLDS)
    for idx, (train_idx, test_idx) in enumerate(cv.split(X, y)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        train_data = Pool(data=X_train, label=y_train, cat_features=cat_feat)
        eval_data = Pool(data=X_test, label=y_test, cat_features=cat_feat)

        model = CatBoostRegressor(**params)
        model.fit(train_data,
                  eval_set=eval_data,
                  early_stopping_rounds=100,
                  verbose=0)

        preds = model.predict(X_test)
        cv_predicts[idx] = mean_absolute_error(y_test, preds)

    return np.mean(cv_predicts)
```

```python
study_cat = optuna.create_study(
    direction="minimize",
    pruner=optuna.pruners.SuccessiveHalvingPruner(),
    study_name="Cat_01")
func = lambda trial: objective_ctb(trial,
                                   X_train,
                                   y_train,
                                   N_FOLDS=train['k_folds'],
                                   random_state=train['random_state'],
                                   cat_feat=cat_features)
study_cat.optimize(func, n_trials=4, show_progress_bar=True)
```

```python
print(f"\tBest value (mae): {study_cat.best_value:.5f}")
print(f"\tBest params:")

for key, value in study_cat.best_params.items():
    print(f"\t\t{key}: {value}")
```
